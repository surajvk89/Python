```python
# Merge

pdfList = ['new_pdf.pdf','newPdf.pdf']

out = 'combined_pdf.pdf'

pdfMerger = PyPDF2.PdfMerger()

for pdf in pdfList:
    pdfMerger.append(pdf)

fobj = open(out,"wb")
pdfMerger.write(fobj)

fobj.close()
```

```python
# split

def splitPdf(pdfFile,ind):
    fObj = open(pdfFile,'rb')
    pdfReader = PyPDF2.PdfReader(fObj)
    
    startInd = 0
    endInd = ind[0]
    
    for i in range(len(ind)+1):
        pdfWriter = PyPDF2.PdfWriter()
        #outPdf = 'outputPdfFile.pdf'
        outPdf = pdfFile.split('.pdf')[0] + '_' + str(i) + '.pdf';
        # combined_pdf_0.pdf
        # combined_pdf_1.pdf
        # combined_pdf_2.pdf
        
        
        for pageInd in range(startInd,endInd):
            pdfWriter.add_page(pdfReader.pages[pageInd])
        oObj = open(outPdf,'wb')
        pdfWriter.write(oObj)
        
        startInd = endInd
        try:
            endInd = ind[i+1]
        except IndexError:
            endInd = len(pdfReader.pages)
        oObj.close()
    
    fObj.close()


def main():
    file = 'combined_pdf.pdf'
    ind = [4,7]
    splitPdf(file,ind)

if __name__ =='__main__':
    main()
```

```python
# watermark

# two pdf files - watermark_pdf, original_pdf


wrObj = open('watermarked_output.pdf','wb')
pdfWriter = PyPDF2.PdfWriter()

wObj = open("watermark.pdf",'rb')
pdfReader = PyPDF2.PdfReader(wObj)

wmpage = pdfReader.pages[0]

oObj = open('original_pdf.pdf','rb')
pdfReader2 = PyPDF2.PdfReader(oObj)
for i in range(len(pdfReader2.pages)):
    OrgPage = pdfReader2.pages[i]
    OrgPage.mergePage(wmpage)
    pdfWriter.add_page(OrgPage)
    
pdfWriter.write(wrObj)
```
